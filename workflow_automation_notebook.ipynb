# File Management

````python
import os
import shutil
from pathlib import Path

FILE_CATEGORIES = {
    "Images": [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff"],
    "Documents": [".pdf", ".docx", ".txt", ".pptx", ".xlsx", ".csv"],
    "Videos": [".mp4", ".mov", ".avi", ".mkv"],
    "Audio": [".mp3", ".wav", ".aac"],
    "Others": []
}

def create_folders(base_path):
    for folder_name in FILE_CATEGORIES.keys():
        folder_path = os.path.join(base_path, folder_name)
        os.makedirs(folder_path, exist_ok=True)

def move_files(base_path):
    for item in os.listdir(base_path):
        item_path = os.path.join(base_path, item)

        if os.path.isdir(item_path):
            continue

        file_ext = Path(item).suffix.lower()
        moved = False

        for folder_name, extensions in FILE_CATEGORIES.items():
            if file_ext in extensions:
                dest_folder = os.path.join(base_path, folder_name)
                shutil.move(item_path, os.path.join(dest_folder, item))
                print(f"Moved {item} to {folder_name}")
                moved = True
                break

        if not moved:
            shutil.move(item_path, os.path.join(base_path, "Others"))
            print(f"Moved {item} to Others")

def organize_directory(directory_path):
    if not os.path.exists(directory_path):
        print("The directory does not exist!")
        return

    create_folders(directory_path)

    move_files(directory_path)

    print("File organization complete!")

if __name__ == "__main__":

    directory_to_organize = input("Enter the path of the directory to organize: ")

    organize_directory(directory_to_organize)
````

## 3. Scheduling a Simple Job
You can use `schedule.every()` with various time intervals to schedule a job.

For example, the following code schedules a job to run every 10 seconds.

````python
def job( ):
    print('Job executed')

# Schedule the job to run every 10 seconds
schedule.every(10).seconds.do(job)
# pending job
````

````python
def job( ):
    print('Job executed')

# Schedule the job to run every 10 seconds
schedule.every(10).seconds.do(job)
# pending job

while True:
    schedule.run_pending()
    time.sleep(1)  # Wait for 1 second before checking for pending jobs again
````

## 4. Running the Scheduled Jobs
Use `schedule.run_pending()` to check for any pending jobs and execute them. You can use a `while` loop to keep the scheduler running.

````python
print('Starting scheduler. Press Ctrl+C to stop.')
try:
    while True:
        schedule.run_pending()
        time.sleep(1)
except KeyboardInterrupt:
    print('Scheduler stopped.')
````

## 5. Scheduling Jobs with Specific Intervals
You can also schedule jobs with specific intervals, like:
- `schedule.every().hour` : Run the job every hour.
- `schedule.every().day.at('10:30')` : Run the job every day at 10:30 AM.
- `schedule.every().monday` : Run the job every Monday.

````python
# Schedule a job to run every minute
schedule.every(1).minute.do(job)
print('Job scheduled to run every minute.')
````

## 6. Example: Scheduling Multiple Jobs
Here's an example of scheduling multiple jobs with different intervals.
One job runs every 10 seconds, while another runs every 30 seconds.

````python
def job1():
    print('Job 1 executed')

def job2():
    print('Job 2 executed')

# Schedule both jobs

schedule.clear() # clear all the jobs

schedule.every(10).seconds.do(job1)
schedule.every(30).seconds.do(job2)

print('Jobs scheduled. Running...')
try:
    while True:
        schedule.run_pending()
        time.sleep(1)
except KeyboardInterrupt:
    print('Scheduler stopped.')
````

## 7. Example: Scheduler with a Countdown Timer
This example shows how to combine a scheduler with a countdown timer function.

````python
def countdown_timer(n):
    while n > 0:
        print(f'Time remaining: {n} seconds')
        time.sleep(1)
        n -= 1
    print('Countdown finished!')

# Schedule the countdown_timer job to run every minute
schedule.clear()
schedule.every().minute.do(countdown_timer, 5)

print('Countdown job scheduled to run every minute.')
try:
    while True:
        schedule.run_pending()
        time.sleep(1)
except KeyboardInterrupt:
    print('Scheduler stopped.')
````

cropped_image = image.crop((left, upper, right, lower))

- `upper`: The y-coordinate of the top edge of the crop box.
- `right`: The x-coordinate of the right edge of the crop box.
- `lower`: The y-coordinate of the bottom edge of the crop box.

````python
# Resize the image for better OCR accuracy
resized_image = gray_image.resize((gray_image.width * 2, gray_image.height * 2))

# Apply binary threshold
threshold_image = resized_image.point(lambda p: p > 128 and 255)  # Convert to binary image (black and white)
threshold_image.show()

# Extract text from the processed image
enhanced_text = pytesseract.image_to_string(threshold_image)
print(f"Enhanced Extracted Text: {enhanced_text}")
````

````python
import smtplib
import ssl
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

sender_email = "automation.practice.madison@gmail.com"
receiver_email = "funnymaddie31@gmail.com"
app_password = test_p

# Create the email
subject = "Test Email from Python"

body = '''
Hi,

This is an automated email sent from Python.

Best,
Madison
'''

# Set up MIME
message = MIMEMultipart()
message["From"] = sender_email
message["To"] = receiver_email
message["Subject"] = subject

# Attach the body with the message
message.attach(MIMEText(body, "plain"))

# Set up the secure SSL context
context = ssl.create_default_context()

# Send the email using Gmail's SMTP server
try:
    with smtplib.SMTP_SSL("smtp.gmail.com", 465, context=context) as server:
        server.login(sender_email, app_password)
        server.sendmail(sender_email, receiver_email, message.as_string())
    print("Email sent successfully!")
except Exception as e:
    print(f"Error: {e}")
````


## Merging Cells

   To merge cells in an Excel worksheet, you need to specify the range of cells you want to merge.

#### Example Code to Merge Cells

```python
from openpyxl import Workbook

# Create a new workbook and select the active sheet
workbook = Workbook()
sheet = workbook.active

# Adding some data to the worksheet
sheet['A1'] = 'Merged Cell Example'

# Merging cells from A1 to C1
sheet.merge_cells('A1:C1')

# Save the workbook to a file
workbook.save('merged_cells_example.xlsx')
```

### Explanation:
- **`sheet.merge_cells('A1:C1')`** merges the range of cells from `A1` to `C1` into a single cell.
- The text in the top-left cell (`A1`) will remain and be centered over the merged range.

#### Unmerging Cells

To unmerge the cells, use the `unmerge_cells()` method:

```python
# Unmerge the cells from A1 to C1
sheet.unmerge_cells('A1:C1')

# Save the workbook to a file after unmerging
workbook.save('unmerged_cells_example.xlsx')
```

### Merging Multiple Ranges

You can merge multiple ranges in the same sheet:

````python
from openpyxl import Workbook

# Create a new workbook and select the active sheet
workbook = Workbook()
sheet = workbook.active

# Merging multiple ranges of cells
sheet.merge_cells('A1:C1')  # Merge A1 to C1
sheet.merge_cells('D2:E3')  # Merge D2 to E3
sheet.merge_cells('B5:D5')  # Merge B5 to D5

# Adding text to the merged cells
sheet['A1'] = 'Merged A1 to C1'
sheet['D2'] = 'Merged D2 to E3'
sheet['B5'] = 'Merged B5 to D5'

# Save the workbook
workbook.save('multi_merged_cells_example.xlsx')
````

## Alignment
### Types of Alignment

- **Horizontal Alignment**: Left, center, right, justify, etc.
- **Vertical Alignment**: Top, middle, bottom, justify, etc.
- **Text Wrapping**: Automatically wrap text if it exceeds the width of the cell.
- **Indentation**: Indent the content of the cell.

### Example: Applying Alignments

Here’s a complete example that demonstrates how to apply horizontal and vertical alignment, wrap text, and indent content.


````python

from openpyxl import Workbook
from openpyxl.styles import Alignment

# Create a new workbook and select the active sheet
workbook = Workbook()
sheet = workbook.active

# Adding data to the worksheet
sheet['A1'] = 'Center Aligned'
sheet['A2'] = 'Right Aligned'
sheet['A3'] = 'Left Aligned'
sheet['A4'] = 'Wrapped Text Example'  # This will wrap text
sheet['A5'] = 'Indented Text'

# Apply different types of alignments
sheet['A1'].alignment = Alignment(horizontal='center', vertical='center')
sheet['A2'].alignment = Alignment(horizontal='right', vertical='bottom')
sheet['A3'].alignment = Alignment(horizontal='left', vertical='top')

# Apply text wrapping (this will wrap the text in the cell)
sheet['A4'].alignment = Alignment(wrap_text=True)

# Apply indentation (the text in this cell will be indented by 2 spaces)
sheet['A5'].alignment = Alignment(indent=2)

# Adjust the column width to see the changes clearly
sheet.column_dimensions['A'].width = 30  # Increase the width of column A

# Save the workbook
workbook.save('alignment_example.xlsx')
````

In ReportLab, a popular Python library for creating PDF documents, you can specify the font size of text using points as the unit of measurement. A point is a typographic unit of measure equal to 1/72nd of an inch.

The height of a line of text in ReportLab depends on several factors beyond just the font size, including leading (the vertical space between lines), line spacing, and any additional padding or margins that are applied. Therefore, the actual height of a line of text may be greater than the specified font size.

On the other hand, the width of a block of text in ReportLab is determined by the number of characters in the text string, the font used, and the width of the container element where the text will be rendered. The width of each character varies depending on the font design, so different font choices at the same point size may result in blocks of text with slightly different widths.

To summarize, while there is a direct correlation between the font size and the height of individual glyphs in a given font, the overall height of a line of text in ReportLab depends on various formatting options like leading and line spacing, whereas the width of a block of text depends primarily on the font choice, the length of the text, and the available container width.

For most purposes, when you specify a font size in ReportLab, you can consider this size as a good approximation of the "height" of your font for single lines of text. If you need to calculate the height of a block of text (especially with multiple lines), you'll need to take into account the leading, which is often 120% of the font size by default but can be adjusted.

Here's a basic way to get the font height:

```font_size = 12  # Example font size
leading = font_size * 1.2  # Default leading in ReportLab is 120% of the font size```


Font size refers to the size of the characters in a font, typically measured in points (pt), pixels (px), ems (em), or percentages (%), depending on the medium or context. Among these, points and pixels are the most common units of measurement.

Points (pt)
Definition: A point is traditionally defined as 1/72 of an inch. This measurement comes from the printing industry and is used to ensure consistency in text size across different printing materials and displays.
Usage: Points are widely used in desktop publishing, graphic design, and standard office applications like word processors and presentation software. They provide a physical reference that helps ensure text appears at a consistent size in printed documents.
Pixels (px)
Definition: A pixel is the smallest unit of a digital image or display that can be controlled. The size of a pixel can vary depending on the resolution and size of the display screen.
Usage: Pixels are common in web design and digital content, where the text size needs to adjust to different screen sizes and resolutions. Using pixels gives designers and developers control over how text appears on various devices.
Ems (em) and Percentages (%)
Definition: Both ems and percentages are relative units. An "em" is based on the font size of the element's parent, while percentages directly relate to the parent element's font size.
Usage: These units are used in web development to create scalable and responsive designs that can adapt to the user's screen size or browser settings.
Understanding Font Size in Context
The concept of font size goes beyond just the height of characters. It also indirectly influences other aspects of how text is displayed, including:

Line Height (Leading): The vertical distance between lines of text, which often scales with the font size to maintain readability.
Ascent and Descent: The ascent is the part of a letter that extends above the baseline (e.g., the top of "d"), while the descent is the part that extends below it (e.g., the bottom of "p"). The font size affects the maximum possible heights of these.
X-height: The height of the lowercase "x" relative to the baseline, which varies between fonts and is a key aspect of how large or small text appears to the reader.

````python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from pdf2image import convert_from_path
from reportlab.lib.colors import red, black, blue, blueviolet, lightgrey, cyan

# Create a canvas, which is required to use the stringWidth function

pdf_file_name = "dummy.pdf"
c = canvas.Canvas(pdf_file_name, pagesize=letter)
page_width, page_height = letter


# Backgroud --- Set the fill color to light grey (or any color of your choice)
c.setFillColor(lightgrey)
# Draw a rectangle that covers the entire page
c.rect(0, 0, page_width, page_height, fill=1, stroke=0)


# header info
# Draw a top rectangle
c.setFillColor(red)
inch = 72
top_rectangle_height = 2*inch
c.roundRect(0,page_height-top_rectangle_height/2,page_width,top_rectangle_height,radius = 30, fill=1, stroke=0)

# keeping track the current y position
current_y_position = page_height-top_rectangle_height/2


# calculating the product box
n_box = 3
left_margin = 72/2 # 1 inch = 72 points
right_margin = 72/2

gap_between_product_box = 10 #n+1 many gaps
page_width, page_hight = c._pagesize

box_width = (page_width - left_margin - right_margin - (n_box+1)* gap_between_product_box)/n_box
box_height = box_width

print(f'box_width = {box_width}')

# how many rows of product we want

n_row = 2

radius = 5

product_name_list = ['Chicken Leg Quarter', 'Beef', 'Pompano', 'Egg', 'Salmon', 'Onion', 'Pompano', 'Egg', 'Chicken Leg Quarter', 'Beef', 'Pompano', 'Egg']
product_price = ['$0.49', '$3.99', '$4.99', '$2.99', '$9.99', '$0.99', '$0.49', '$3.99', '$4.99', '$2.99', '$9.99', '$0.99']

# main box
for i in range(0,n_row):
  y = current_y_position - box_height- gap_between_product_box
  x = left_margin+gap_between_product_box
  for j in range(0,n_box):
    product_name_index = n_box*i+j
    main_product_box(c, x, y, box_width, box_height,product_name_list[product_name_index],product_price[product_name_index] )
    x+=box_width+ gap_between_product_box
  current_y_position = y


## Vertical Text

# Define your text and rotation angle
text = 'Vertical text'
angle = 90  # Adjust for desired rotation
# Save current canvas state
c.saveState()
# Rotate the canvas
c.rotate(angle)
# c.setFillColorRGB(255,255,255) #white
c.setFillColor(white)
c.setFontSize(30)
# Draw the text with adjusted coordinates (optional for positioning)
c.drawString(current_y_position, -30, text)  # Adjust coordinates if needed

# Restore the canvas state
c.restoreState()


# Footer
# Text to be centered
text = "Designed by: Name \t www.website.com \t 210-123-4567"

# Calculate the center position
text_width = c.stringWidth(text)
center_x = (c._pagesize[0] - text_width) / 2

# Draw the text centered horizontally
c.drawString(center_x, 10, text)

# save the canvas
c.showPage()  #end of the page, it will help us to create multiple pages
c.save()

#Convert PDF to image(s)
images = convert_from_path(pdf_file_name)

# display the first-page
display(images[0])

````
